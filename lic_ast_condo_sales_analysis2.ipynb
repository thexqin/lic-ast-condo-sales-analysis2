```python
# @title load

import pandas as pd
import numpy as np
import requests
import json
from IPython.display import display, Markdown

df = pd.read_csv('23-15sub.csv', parse_dates=['SALE DATE'])
```

```python
df = df[df.duplicated('aptno', False)].sort_values(['aptno', 'SALE DATE'], ascending=[True, False]).reset_index(drop=True).copy()
df
```

```python
df = df[df.aptno != 404].reset_index(drop=True).copy()
df.head()
```

```python
for i in df.aptno.unique():
    check(i)
```

```python
df['changeM'] = np.nan
df['change$'] = np.nan
df['change%'] = np.nan

for i in range(0, len(df), 2):
    temp = df.loc[i, 'SALE DATE'] - df.loc[i+1, 'SALE DATE']
    df.loc[i, 'changeM'] = temp.days / 30
    df.loc[i, 'change$'] = df.loc[i, 'SALE PRICE'] - df.loc[i+1, 'SALE PRICE']
    df.loc[i, 'change%'] = df.loc[i, 'SALE PRICE'] / df.loc[i+1, 'SALE PRICE']

df
```

```python
df['change%'].describe()
```

```python
df[df['change%'] > 0.8].boxplot('change%', vert=False, grid=False, figsize=(10, 2))
```
